```python
import pandas as pd
```

```python
df = pd.read_csv('cyano_lysate_experiment.csv')
df.drop(columns=['time'], inplace=True)
df
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6)) 

sns.heatmap(df, annot=True, fmt=".4f", cmap="Blues")

plt.title("OD600 Heatmap")
plt.xlabel("Day")
plt.ylabel("Sample")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Melt the DataFrame to long format
df_long = df.melt(id_vars=["Sample"], var_name="Day", value_name="OD600")

# Plot using seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_long, x="Day", y="OD600", marker="o")
plt.title("OD600 Over Time")
plt.xlabel("Day")
plt.ylabel("OD600")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
